```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.dates as dates
```

```python
def rain(total, mean, days):
    # total amount of rain over season, mean amount per rainfall event, duration of growing season
    meanDays = total / mean / days # alpha in exponential distribution
    randomEvents = [0]*days # initialize empty array to represent whether each day is rainy
    
    # find days between rainfall events
    i = 0
    while i < days:
        k = round(np.random.exponential(scale = 1/meanDays, size = None))
        if(i+k > days-1):
            break
        else:
            randomEvents[i+k] = 1
            i+=k
    
    # find total number of rainy days
    i = 0
    randRainyDays = 0
    while i < days:
        if (randomEvents[i] == 1):
            randRainyDays += 1
        i += 1
    # print('There are', randRainyDays, 'rainy days in the growing season.')
    
    # randomize amount of rainfall for each rainy day based on exponential distribution 
    randomRain = np.random.exponential(mean, randRainyDays) 
    # print(randomRain)
    
    i = 0
    j = 0
    totalRandRain = [0] * days # initialize empty array to represent all rainfall events
    while i < days:
        if (randomEvents[i] == 1): # if a rainy day
            totalRandRain[i] = randomRain[j] # the rainy day gets assigned to a random rainfall depth
            j += 1
            i += 1
        else: # if a dry day, the depth remains 0
            totalRandRain[i] = 0 
            i += 1
    # print(totalRandRain)

    rainfall = totalRandRain # rainfall in mm

    return rainfall
```

```python
def s_t(rainfall, soil_texture, Z, bio=False):
    # Input rainfall array from rain function, desired soil texture, active depth (can be variable with crop), 
    # presence of biochar amendment
    
    # Returns arrays of soil moisture values for [without irrigation, with, irrigation amount]
    
    # initialize soil parameters based on texture
    if soil_texture=='s': # sand
        Ks = 200 # in reality, K >200 cm/d
        B = 12.1
        sh = 0.08
        sw = 0.11 
        sx = 0.33
        sfc = 0.35
        n = 0.35
        
        s = [sfc]
        s_micro = np.zeros(180)
        s_micro[0] = sfc
        
        if bio==True:
            Ks = Ks - 0.921*Ks
            sh = sh - 0.1018*sh
            sw = sw - 0.09355*sw
            sfc = sfc + 0.2945*sfc
            n = n + 0.03473*n
            sx = sx - 0.003738*sx
    elif soil_texture=='ls': # loamy sand
        Ks = 100
        B = 12.7
        sh = 0.08
        sw = 0.11
        sx = 0.31
        sfc = 0.52
        n = 0.42
        
        s = [sfc]
        s_micro = np.zeros(180)
        s_micro[0] = sfc
        
        if bio==True:
            Ks = Ks - 0.2986*Ks
            sh = sh + 0.1624*sh
            sw = sw - 0.09533*sw
            sfc = sfc + 0.6458*sfc
            n = n + 0.02645*n
            sx = sx + 0.5969*sx
            
    elif soil_texture=='sl': # sandy loam
        Ks = 80
        B = 13.8
        sh = 0.14
        sw = 0.18
        sx = 0.46
        sfc = 0.56
        n = 0.43
        
        s = [sfc]
        s_micro = np.zeros(180)
        s_micro[0] = sfc
    
        if bio==True:
            n = n + 0.09203*n
            Ks = Ks - 0.451*Ks
            sh = sh - 0.06453*sh
            sw = sw - 0.07140*sw
            sfc = sfc + 0.1691*sfc
            sx = sx + 0.2317*sx 
            
    elif soil_texture=='l': # loam
        Ks = 20
        B = 14.8
        sh = 0.19
        sw = 0.24
        sx = 0.57
        sfc = 0.65
        n = 0.45
        
        s = [sfc]
        s_micro = np.zeros(180)
        s_micro[0] = sfc
    
        if bio==True:
            n = n + 0.06468*n
            Ks = Ks - 0.006289*Ks
            sh = sh - 0.06685*sh
            sw = sw + 0.1399*sw
            sfc = sfc + 0.1383*sfc
            sx = sx + 0.2087*sx
            
    else: # clay loam
        Ks = 10 # in reality, <10
        B = 26.8
        sh = 0.47
        sw = 0.52
        sx = 0.78
        sfc = 1
        n = 0.5
        
        s = [sfc]
        s_micro = np.zeros(180)
        s_micro[0] = sfc
    
        if bio==True:
            n = n + 0.007543*n
            Ks = Ks + 2.75*Ks
            sh = sh - 0.06685*sh
            sw = sw - 0.05890*sw
            sfc = sfc + 0.1472*sfc
            sx = sx + 0.1394*sx
        
    
    
#     # biochar effect
#     if bio==True:
#         # ENABLE FOR W/ BIOCHAR
#         sh = sh - 0.06685*sh
#         sw = sw - 0.07311*sw
#         # sx = sx + 0.2111*sx
#         sfc = sfc + 0.1745*sfc

#         Ks = Ks + 0.1140*Ks
        
#         n = n + 0.06468*n
    
    # precipitation minus canopy interception 
    p = []
    i = 0
    l = n * Z
    while i < 180:
        if((rainfall[i])>0):
            p.append((rainfall[i])/l)
            i+=1
        else:
            p.append(0)
            i+=1
    
    # soil moisture equation
    Ew = .1 ##mm/day

    i = 1        

    Emax = 5.5 # mm/day 
    dsdt = sfc - sx
    
    s_irr = np.zeros(180)

    # without irrigation
    while i < 180:
        if s[i-1] > 1:
            s[i-1] = 1
        if s[i-1] > 0 and s[i-1] <= sh :
            s.append(s[i-1]+ p[i-1])
        elif s[i-1] > sh and s[i-1] <= sw :
            s.append(s[i-1]+ p[i-1] - ((Ew*((s[i-1]-sh)/(sw-sh)))/l))
        elif s[i-1] > sw and s[i-1] <= sx:
            s.append(s[i-1] + p[i-1] - ((Ew + ((Emax - Ew)*((s[i-1]-sw)/(sx-sw)))))/l)
        elif s[i-1] > sx and s[i-1] <= sfc:
            s.append(s[i-1] + p[i-1] - (Emax/l))
        elif s[i-1] > sfc*1.1:
            s.append(sfc*1.1)
        elif s[i-1] > sfc and s[i-1] <= 1:
            s.append(s[i-1] + p[i-1] -(Emax/l)-((Ks/(np.exp(B*(1-sfc))-1))*((np.exp(B*(s[i-1]-sfc))-1)/l)))
        i+=1
        
    # soil moisture with microirrigation
    i = 1
    while i < 180:
        if s_micro[i-1] > 1:
            s_micro[i-1] = 1
        if s_micro[i-1] > 0 and s_micro[i-1] <= sh :
            s_micro[i] = (s_micro[i-1]+ p[i-1])
        elif s_micro[i-1] > sh and s_micro[i-1] <= sw :
            s_micro[i] = (s_micro[i-1]+ p[i-1] - ((Ew*((s_micro[i-1]-sh)/(sw-sh)))/l))
        elif s_micro[i-1] > sw and s_micro[i-1] <= sx:
            # add water equivalent to losses
            s_irr[i] = Emax
            irr = (s_irr[i] / l)
            s_micro[i] = sx + p[i-1] # + irr - (Emax/l)
            s_micro[i-1] = sx
        elif s_micro[i-1] > sx and s_micro[i-1] <= sfc:
            s_micro[i] = (s_micro[i-1] + p[i-1] - (Emax/l))
        elif s_micro[i-1] > sfc*1.1:
            s_micro[i] = (sfc*1.1)
        elif s_micro[i-1] > sfc and s_micro[i-1] <= 1:
            s_micro[i] = (s_micro[i-1] + p[i-1] -(Emax/l)-((Ks/(np.exp(B*(1-sfc))-1))*((np.exp(B*(s[i-1]-sfc))-1)/l)))
        i += 1
        
    return [s, s_micro, s_irr]
```
